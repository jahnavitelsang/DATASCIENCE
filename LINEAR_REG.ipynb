# **MULTIPLE LINEAR REGRESSION**

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('/content/Cars_4vars.csv')
```

```python
df.head()
```

```python
X_model1 = df[['HP']]
y_mpg = df['MPG']
```

```python
model1 = LinearRegression()
model1
```

```python
model1.fit(X_model1, y_mpg)
```

```python
y_pred = model1.predict(X_model1)
y_pred
```

```python
r2 = r2_score(y_mpg, y_pred)
r2
```

```python
rmse = np.sqrt(mean_squared_error(y_mpg, y_pred))
rmse
```

```python
X_model2 = df[['HP','SP']]
y_mpg = df['MPG']
model2 = LinearRegression()
model2.fit(X_model2, y_mpg)
y_pred2 = model2.predict(X_model2)
r2 = r2_score(y_mpg, y_pred2)
print(r2)
rmse = np.sqrt(mean_squared_error(y_mpg, y_pred2))
rmse
```

```python
X_model3 = df[['HP','SP','VOL']]
y_mpg = df['MPG']
model3 = LinearRegression()
model3.fit(X_model3, y_mpg)
y_pred3 = model3.predict(X_model3)
r2 = r2_score(y_mpg, y_pred3)
print(r2)
rmse = np.sqrt(mean_squared_error(y_mpg, y_pred3))
rmse
```

```python
X_model4 = df[['HP','SP','VOL','WT']]
y_mpg = df['MPG']
model4 = LinearRegression()
model4.fit(X_model4, y_mpg)
y_pred4 = model4.predict(X_model4)
r2 = r2_score(y_mpg, y_pred4)
print(r2)
rmse = np.sqrt(mean_squared_error(y_mpg, y_pred4))
rmse
```

```python
# see that after the vol , there is no chnage in the r2 and rmse so we can stop at vol #
```

```python
vif_hp_sp = 1 / (1- (r2))   # used to detect the multicollinearuty #
vif_hp_sp
```

```python
print("vif value :", vif_hp_sp)
print("value of r2:",r2)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize = (8,5))    # 8 and 6 are the width and height of the plot#
sns.heatmap(df.corr(numeric_only = True), annot=True , cmap = 'coolwarm')

#df.corr() calculates the Pearson correlation coefficient between all pairs of columns in your dataset.#
# only corelation between numbers needed and #
#annot=True is the number are written inside the box#
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x="Clicked on Ad", y="Daily Time Spent on Site", data = df)
plt.title('Daily Time Spent on Site vs Clicked on Ad')
plt.xlabel('Clicked on Ad')
plt.ylabel('Daily Time Spent on Site')
plt.show()
# many outliers#
```

convert timestamp to data nd time and extract  new features

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'], format='%d-%m-%Y %H:%M')
#tells Pandas to look at the strings in the Timestamp column and convert them in datetime objects#
df['Hour'] = df['Timestamp'].dt.hour
df['Day'] = df['Timestamp'].dt.day
```

```python
df[['Timestamp', 'Hour', 'Day']].head()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x ='Hour', hue = 'Clicked on Ad' ,data=df)
plt.title('Hourly Distribution')
plt.xlabel('Hour')
plt.ylabel('Count')

plt.show()
```

```python
X = df[['Daily Time Spent on Site', 'Age', 'Area Income', 'Daily Internet Usage', 'Male', 'Hour','Day']]
y = df['Clicked on Ad']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, recall_score , precision_score , f1_score
```

```python
#initialize and fit the modem 1st#

logmodel = LogisticRegression()
logmodel.fit(X,y)
```

```python
y_pred = logmodel.predict(X)
```

```python
conf_metrics = confusion_matrix(y,y_pred)
conf_metrics
```

```python
accuracy = accuracy_score(y,y_pred)
accuracy
```

```python
classification_rep = classification_report(y,y_pred)
print(classification_rep)
```

```python
recall = recall_score(y,y_pred)
recall
```

```python
precision = precision_score(y,y_pred)
precision
```

```python
f1_score = f1_score(y,y_pred)
f1_score
```

```python
#specificity that is tp/tp+fn#

tn,fp,fn,tp = confusion_matrix(y,y_pred).ravel()
specificity = tn / (tn+fp)
print(specificity)
```

```python
from sklearn.metrics import roc_curve,auc
y_pred = logmodel.predict(X)
y_pred_prob = logmodel.predict_proba(X)[:,1]

#roc and auc curve#                 # plot the auc roc ....for that we 1st need the x and y rows and columns and then we need the curve #
fpr,tpr,thresholds = roc_curve(y,y_pred_prob)
roc_auc = auc(fpr,tpr)


#Plot teh curve #
plt.figure(figsize=(6, 4))
plt.plot(fpr, tpr, color='red', lw=1, label='ROC curve (area = %0.2f)' % roc_auc)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```
