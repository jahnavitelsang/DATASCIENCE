```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans 
import pandas as pd 
import numpy as np
```

```python
iris=datasets.load_iris()
iris.data
```

```python
iris.target
```

```python
iris.target_names
```

```python
X=pd.DataFrame(iris.data)
X
```

```python
X.columns=['Sepal_Length','Sepal_width','Petal_Length','Petal_width']
y=pd.DataFrame(iris.target)
y.columns=['Targets']
```

```python
plt.figure(figsize=(14,7))
colormap=np.array(['red','lime','black','blue'])
```

```python
plt.subplot(1,2,1)
plt.scatter(X.Sepal_Length,X.Sepal_width,c=colormap[y.Targets],s=40)
plt.title('Sepal')
plt.subplot(1,2,2)
plt.scatter(X.Petal_Length,X.Petal_width,c=colormap[y.Targets],s=40)
plt.title('Petal')
```

```python
#build the model
#K means cluster 
model=KMeans(n_clusters=3)
model.fit(X)
predY=model.labels_
```

```python
plt.figure(figsize=(14,7))
colormap=np.array(['red','lime','black','blue'])
plt.subplot(1,2,1)
plt.scatter(X.Petal_Length,X.Petal_width,c=colormap[y.Targets],s=40)
plt.title('Real Classification')
plt.subplot(1,2,2)
plt.scatter(X.Petal_Length,X.Petal_width,c=colormap[model.labels_],s=40)
plt.title('K mean classification')
```

```python
#method to find number of clusters using elbow method 
SSE=[]
for k in range(2,11):
  model=KMeans(n_clusters=k)
  model.fit(X)
  sse_score=model.inertia_
  SSE.append(sse_score)

#plt.style("fivethirtyeight")
plt.plot(range(2,11),SSE)
plt.xticks(range(2,11))
plt.xlabel('Number of Clusters')
plt.ylabel('SSE')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

H_clustering = AgglomerativeClustering(n_clusters=3)
H_clustering.fit(X)
predY = H_clustering.labels_
predY
```

```python
#View the results
# Set the size of the plot
plt.figure(figsize=(14,7))
 
# Create a colormap
colormap = np.array(['red', 'lime', 'black','blue'])
 
# Plot the Original Classifications
plt.subplot(1, 2, 1)
plt.scatter(X.Petal_Length, X.Petal_width, c=colormap[y.Targets], s=40)
plt.title('Real Classification')
 
# Plot the Models Classifications
plt.subplot(1, 2, 2)
predY = H_clustering.labels_
#predY = np.choose(predY, [1, 0, 2]).astype(np.int64)
plt.scatter(X.Petal_Length, X.Petal_width, c=colormap[predY], s=40)
plt.title('Hierarchical clustering')
```

```python
import sklearn.metrics as sm
for n in range(2,11):
    H_clustering = AgglomerativeClustering(n_clusters=n)
    H_clustering.fit(X)
    print(n, sm.silhouette_score(X, H_clustering.labels_))
```

```python
sm.silhouette_score(X, predY)
```

```python
# A list holds the silhouette coefficients for each k
silhouette_coefficients = []
for k in range(2, 11):
    model = KMeans(n_clusters=k)
    model.fit(X)
    score = sm.silhouette_score(X, model.labels_)
    silhouette_coefficients.append(score)
#plt.style.use("fivethirtyeight")
plt.plot(range(2, 11), silhouette_coefficients)
plt.xticks(range(2, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Coefficient")
plt.show()
```

```python
# A list holds the silhouette coefficients for each k
SSE = []
for k in range(2, 11):
    model = KMeans(n_clusters=k)
    model.fit(X)
    sse_score = model.inertia_
    SSE.append(sse_score)
#plt.style.use("fivethirtyeight")
plt.plot(range(2, 11), SSE)
plt.xticks(range(2, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```
