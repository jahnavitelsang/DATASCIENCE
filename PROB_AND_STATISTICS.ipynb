```python
pop_mean = 1000
pop_std = 50
alpha = 0.05
```

```python
#firstly given is pop mean and pop sdt, need to take the alpha valuse as default. #
```

```python
sample_data = df['Lifespan (Hours)']   #to determone the z statistic we need sample data and its mean and also the n value that is the length of sample data#
sample_mean = sample_data.mean()
n = len(sample_data)
```

```python
z_statistic = (sample_mean - pop_mean)/(pop_std/np.sqrt(n))
```

```python
s = norm(1000,50)
p_value_twotailed = 2*(1-norm.cdf(abs(z_statistic)))
```

```python
sample_mean
```

```python
z_statistic
```

```python
np.round(p_value_twotailed,4)
```

```python
if p_value_twotailed > alpha:
  print("Accept the null hypothesis")
else:
  print("Reject the null hypothesis")
```

```python
lungcap = df['LungCap']    # to get the sample mean , need sample dataset and also the n value#
n = len(lungcap)
n
```

```python
sample_mean = lungcap.mean()
sample_std = lungcap.std()
```

```python
# 95% confidence
```

```python
from scipy.stats import t
confidence = 0.95
t_crit = t.ppf((1+confidence)/2,df = n-1)
```

```python
margin_of_error = t_crit * sample_std/np.sqrt(n)
```

```python
ci_upper = sample_mean + margin_of_error
ci_lower = sample_mean - margin_of_error
```

```python
print("lower bound is :" ,np.round(ci_lower,2))
print("upper bound is :" ,np.round(ci_upper,2))
```
